```python
import os
import json
import pandas
from PIL import Image
from tqdm.notebook import tqdm
```

```python
def webp2png(img, newImg):
    with Image.open(img) as im:
        im.save(newImg)

def findAllwebp(folder):
    images = []

    for subdir, _, files in os.walk(folder):
        for file in files:
            ext = os.path.splitext(file)[1]
            if ext == ".webp":
                images.append(os.path.join(subdir, file))
    
    return images

for img in tqdm(findAllwebp("../../SchaleDB-adjusted")):
    webp2png(img, img[:-4]+'png')
    os.remove(img)
```

```python
studentDataFile = "../../SchaleDB/data/jp/students.json"
```

```python
with open(studentDataFile, "r", encoding="utf-8") as f:
    data = json.load(f)
f.close()
```

```python
allIcons = set()
for student in data:
    skillList = student['Skills']
    for skill in skillList:
        if skill['SkillType'] != 'autoattack':
            allIcons.add(skill['Icon'])
```
